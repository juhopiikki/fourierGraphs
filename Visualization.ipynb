# Draw some cool visualizations

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')
```

```python
# Fourier series
# P period
# N seasonality
# t timeline

def fourier_sum_a(P, N, t):
    sum = np.zeros((N,P))
    for i in np.arange(N):
        sum[i] = np.cos((2*np.pi*(i+1)*t)/P)
    return sum
    
def fourier_sum_b(P, N, t):
    sum = np.zeros((N,P))
    for i in np.arange(N):
        sum[i] += np.sin((2*np.pi*(i+1)*t)/P)
    return sum
```

```python
# Create base series
P = 300
N = 5
xs = np.arange(0,P,1)
f_a = fourier_sum_a(P, N, xs)
f_b = fourier_sum_b(P, N, xs)
```

```python
# Run again to create new plot with different random coefficients
plt.figure(figsize=(15, 3.8), dpi=160)
plt.rcParams['axes.facecolor'] = 'black'

for j in range(6):
    # Create random coefficients. Adjust loc and scale to different results. 
    # loc=1 => closer to base series, scale adjust the variation of the different series.
    loc = 0.0 # -0.2
    scale = 0.3
    a = np.random.normal(loc, scale, N)
    b = np.random.normal(loc, scale, N)

    # multiply and sum up
    mult_a = np.dot(np.transpose(f_a), a)
    mult_b = np.dot(np.transpose(f_b), b)

    # Try different configurations and scales
    # turkoosi to green
    red = 0.1+j*0.02
    green = 0.2+j*0.15
    blue = 0.3-j*0.02
    
    # green
    #red = 0.1+j*0.02
    #green = 0.1+j*0.18
    #blue = 0.1+j*0.01
    
    # violet to red:
    #red = 0.15+j*0.17
    #green = 0.1+j*0.05
    #blue = 0.45-j*0.01
    color = (red, green, blue)
    plt.plot(xs, mult_a + mult_b, color=color, linewidth=1.7)# +j*0.5)

plt.xticks([])
plt.xlim(0,299)
plt.ylim(-4.0,4.0)
plt.yticks([])
plt.savefig('images/black_fourier21.png', dpi=300)
plt.show()
```
